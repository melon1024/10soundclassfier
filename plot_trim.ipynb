```python
import librosa.util
import librosa.effects
import librosa
import matplotlib.pyplot as plt
%matplotlib nbagg
import librosa.display
import numpy as np
```

```python
y, sr = librosa.load('/home/sovis2016/asd/7/15419__pagancow__dorm-door-opening.wav',
                     sr=None, duration=30, offset=30)
```

```python
#y = librosa.util.pad_center(y, size=len(y) + 5 * sr)
```

```python
hop_length = 512

intervals = librosa.effects.split(y, hop_length=hop_length)

interval_times = librosa.samples_to_time(intervals, sr=sr)
```

```python
interval_times
```

```python
plt.figure()
librosa.display.waveplot(y, sr=sr)

for ival in interval_times:
    plt.axvspan(ival[0], ival[1], ymin=-1, ymax=1, alpha=0.25, color='r')
```
